```python
RR=np.zeros([150])
for ii in range(1000):
    for jj in range(1000):
        if jj>ii:
            r = ((datarand[0,ii]-datarand[0,jj])**2+(datarand[1,ii]-datarand[1,jj])**2)**0.5
            ll = np.int(np.floor(r))
            RR[ll]+=1
            
DD1=np.zeros([150])
for ii in range(1000):
    for jj in range(1000):
        if jj>ii:
            r = ((data1[0,ii]-data1[0,jj])**2+(data1[1,ii]-data1[1,jj])**2)**0.5
            ll = np.int(np.floor(r))
            DD1[ll]+=1

DD2=np.zeros([150])
for ii in range(1000):
    for jj in range(1000):
        if jj>ii:
            r = ((data2[0,ii]-data2[0,jj])**2+(data2[1,ii]-data2[1,jj])**2)**0.5
            ll = np.int(np.floor(r))
            DD2[ll]+=1
            
DR1=np.zeros([150])
for ii in range(1000):
    for jj in range(1000):
        #if jj>ii:
            r = ((datarand[0,ii]-data1[0,jj])**2+(datarand[1,ii]-data1[1,jj])**2)**0.5
            ll = np.int(np.floor(r))
            DR1[ll]+=1
            
DR2=np.zeros([150])
for ii in range(1000):
    for jj in range(1000):
        #if jj>ii:
            r = ((datarand[0,ii]-data2[0,jj])**2+(datarand[1,ii]-data2[1,jj])**2)**0.5
            ll = np.int(np.floor(r))
            DR2[ll]+=1
            
```

```python
s1=DD1/RR-1
plt.plot(np.linspace(1,150,150),s1)
```

```python
s2=DD2/RR-1
plt.plot(np.linspace(1,150,150),s2)
```

```python
s3=(DD1-DR1+RR)/RR
plt.plot(np.linspace(1,150,150),s3)
```

```python
s4=(DD2-DR2+RR)/RR
plt.plot(np.linspace(1,150,150),s4)
```

# Random Circles

```python
import numpy as np
import matplotlib.pyplot as plt
R = np.random.rand(2,1000)*100
plt.scatter(R[0,:],R[1,:],marker='.')
```

```python
num_points=1000
num_circles=20
radius=30


p_per_circle=num_points/num_circles
num_points2=num_points+num_circles



D1=np.zeros([2,num_points])
r1=np.random.rand(2,num_circles)*100
k=0
for ii in range(num_circles):
    for jj in range(p_per_circle):
        theta=np.random.rand(1)*2*np.pi
        x=radius*np.sin(theta)+r1[0,ii]
        y=radius*np.cos(theta)+r1[1,ii]
        while x>100. or x<0. or y>100. or y<0.:
            theta=np.random.rand(1)*2*np.pi
            x=radius*np.sin(theta)+r1[0,ii]
            y=radius*np.cos(theta)+r1[1,ii]
        D1[0,k]=x
        D1[1,k]=y
        k+=1
        
D2=np.zeros([2,num_points2])
r2=np.random.rand(2,num_circles)*100
l=0+num_circles
for jj in range(l):
    D2[0,jj]=r2[0,jj]
    D2[1,jj]=r2[1,jj]
for ii in range(num_circles):
    for jj in range(p_per_circle):
        theta=np.random.rand(1)*2*np.pi
        x=radius*np.sin(theta)+r2[0,ii]
        y=radius*np.cos(theta)+r2[1,ii]
        while x>100. or x<0. or y>100. or y<0.:
            theta=np.random.rand(1)*2*np.pi
            x=radius*np.sin(theta)+r2[0,ii]
            y=radius*np.cos(theta)+r2[1,ii]
        D2[0,l]=x
        D2[1,l]=y
        l+=1
        
```

```python
plt.gca().set_aspect('equal')
plt.scatter(D1[0,:],D1[1,:],marker='.')
```

```python
plt.gca().set_aspect('equal')
plt.scatter(D2[0,:],D2[1,:], marker='.')
```

```python
RR=np.zeros([150])
for ii in range(1000):
    for jj in range(1000):
        if jj>ii:
            r = ((R[0,ii]-R[0,jj])**2+(R[1,ii]-R[1,jj])**2)**0.5
            ll = np.int(np.floor(r))
            RR[ll]+=1
            
DD1=np.zeros([150])
for ii in range(1000):
    for jj in range(1000):
        if jj>ii:
            r = ((D1[0,ii]-D1[0,jj])**2+(D1[1,ii]-D1[1,jj])**2)**0.5
            ll = np.int(np.floor(r))
            DD1[ll]+=1

DD2=np.zeros([150])
for ii in range(1020):
    for jj in range(1020):
        if jj>ii:
            r = ((D2[0,ii]-D2[0,jj])**2+(D2[1,ii]-D2[1,jj])**2)**0.5
            ll = np.int(np.floor(r))
            DD2[ll]+=1
```
